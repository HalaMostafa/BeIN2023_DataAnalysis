# Joins:

```python
import pandas as pd
```

```python
df_1 = pd.DataFrame({'student_id': {0: 'a', 1: 'b', 2: 'c',3:'e'},
                   'arabic': {0: 1, 1: 3, 2: 5,3:2},})
df_1
```

```python
df_2 = pd.DataFrame({'student_id': {0: 'a', 1: 'b', 2: 'c',3:'d'},
                   'english': {0: 1, 1: 3, 2: 5,3:6},})
df_2
```

### LEFT JOIN:

```python
left_join_df = pd.merge(df_1,df_2,how='left',on='student_id').fillna(0)
left_join_df
```

### RIGHT JOIN

```python
right_join_df = pd.merge(df_1,df_2,how='right',on='student_id').fillna(0)
right_join_df
```

### INNER JOIN:

```python
inner_join_df = pd.merge(df_1,df_2,on='student_id')
inner_join_df
```

### OUTER JOIN:

```python
outer_join_df = pd.merge(df_1,df_2,how='outer',on='student_id').fillna(0)
outer_join_df
```

```python
left = pd.DataFrame(
   {
      "key1": ["K0", "K0", "K1", "K2"],
      "key2": ["K0", "K1", "K0", "K1"],
      "A": ["A0", "A1", "A2", "A3"],
      "B": ["B0", "B1", "B2", "B3"],
   }
)
left
```

```python
right = pd.DataFrame(
   {
      "key1": ["K0", "K1", "K1", "K2"],
      "key2": ["K0", "K0", "K0", "K0"],
      "C": ["C0", "C1", "C2", "C3"],
      "D": ["D0", "D1", "D2", "D3"],
   }
)
right
```

```python
#left join:
fatima = pd.merge(left, right, how= "left", on=["key1", 'key2'])
fatima
```

```python
# inner join
mohamed = pd.merge(left,right , how="inner",on= ["key1","key2"])
mohamed
```

```python
# outer join
Zaynab =pd.merge(left, right, how="outer", on=["key1","key2"])
Zaynab
```

### suffix on merge

```python
leftDf = pd.DataFrame({'student_id': {0: 'a', 1: 'b', 2: 'c',3:'d'},
                   'english': {0: 1, 1: 4, 2: 5,3:6},})
leftDf
```

```python
rightDf = pd.DataFrame({'student_id': {0: 'a', 1: 'b', 2: 'c',3:'d'},
                   'english': {0: 5, 1: 3, 2: 5,3:6},
                   'arabic': {0: 5, 1: 3, 2: 7,3:8},})
rightDf
```

```python
mergedDf = pd.merge(leftDf,rightDf,how = 'outer',on='student_id',suffixes=('_left', '_right'))
mergedDf
```

```python
# check differences between values of english_x vs english_y

mergedDf["diff"] = mergedDf["english_left"] - mergedDf["english_right"]
mergedDf["diff_2"] = mergedDf["diff"].astype(bool)
number_of_different_rows = mergedDf['diff_2'].sum()
number_of_different_rows
# total_students = mergedDf["student_id"].value_counts().sum()
total_students = mergedDf["student_id"].count()
total_students
precentage_of_diff = (number_of_different_rows/total_students)*100
precentage_of_diff
```

# concate

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    # index=[0, 1, 2, 3],

)
df1
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    # index=[4, 5, 6, 7],
)
df2
```

```python
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    # index=[8, 9, 10, 11],
)
df3
```

```python
# concat df1 ,df2 and df3 rows 
concatDF = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
concatDF
```
